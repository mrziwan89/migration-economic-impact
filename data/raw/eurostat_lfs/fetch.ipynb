```python
import pandas as pd
import os
# Create the directory if it doesn't exist
output_dir = '../data/raw/eurostat_lfs'
os.makedirs(output_dir, exist_ok=True)
# Eurostat API base URL for SDMX-CSV format
base_url = "https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/"
format_param = "?format=SDMX-CSV&compressed=false"
# The 3 Datasets we need
datasets = {
    "lfsa_etpgan": "Temporary employees by sex, age and citizenship", # For Pillar 2
    "lfsa_eegacob": "Employment by education and citizenship",        # For Pillar 2
    "ilc_peps05": "At-risk-of-poverty rate by citizenship",            # For Pillar 3
    "lfsa_ergaedn": "Employment by education and citizenship"
}
for code, description in datasets.items():
    print(f"Downloading: {description} ({code})...")
    url = f"{base_url}{code}{format_param}"
    
    try:
        # Read directly from the Eurostat API
        df = pd.read_csv(url)
        
        # Save to your raw data folder
        file_path = os.path.join(output_dir, f"{code}.csv")
        df.to_csv(file_path, index=False)
        print(f"✅ Saved to {file_path}")
    except Exception as e:
        print(f"❌ Failed to download {code}. Error: {e}")
print("Eurostat extraction complete!")
```
